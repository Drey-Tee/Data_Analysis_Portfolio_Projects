```python
#converting the budget and gross columns to int64
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
#reading the dataframe
df
```

```python
#using regular expressions to formate date
df['format_released'] = pd.to_datetime(df['released'].str.extract(r'(\w+ \d+, \d+)', expand=False), format='%B %d, %Y')
df
```

```python
#creating a new column to get accurate year
df['correctyear'] = df['format_released'].astype(str).str[:4]
df
```

```python
#sorting gross values by descending order
df.sort_values(by=['gross'],inplace = False, ascending = False)
```

```python
#dropping all dupplicates in the dataframe
df.drop_duplicates()
```

```python
#dropping all n/a valaues in the dataframe
df.dropna()
```

```python
# Plot budget vs gross using plotly
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#correlation matrix
df.corr(method= 'pearson')
```

```python
#correlation matrix for numeric featires
corr_matrix = df.corr(method= 'pearson')
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at company - nonnumeric correlation
# Numerise companies
df_numerised = df
for col_name in df_numerised.columns:
    if (df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes
        
df_numerised
```

```python
# Look at the correlation among every attribute
correlation_matrix = df_numerised.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at the correlation by pairing up every attribute
correlation_mat = df_numerised.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# Pair up attributes in ascending order of correlation index
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Filter the highest correlation pairs
high_corr = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1)]
high_corr
```
